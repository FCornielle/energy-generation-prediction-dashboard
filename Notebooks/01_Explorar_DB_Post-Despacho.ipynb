#### Importar y Unir Datos de Generaci√≥n de los Archivos .parquet

```python
import os
import glob
import pandas as pd

# Carpeta que contiene los archivos .parquet
folder_path = r"..\data\raw"

# Encontrar todos los archivos .parquet en la carpeta
parquet_files = glob.glob(os.path.join(folder_path, "*.parquet"))

# Lista para almacenar los DataFrames
df_list = []

# Cargar cada archivo Parquet y a√±adirlo a la lista
for file in parquet_files:
    print(f"Importando {file}...")
    df_temp = pd.read_parquet(file)
    df_list.append(df_temp)

# Concatenar todos los DataFrames en uno solo
if df_list:
    df_merged = pd.concat(df_list, ignore_index=True)
    # Guardar el DataFrame combinado en un nuevo archivo Parquet
    output_path = r"..\Data\interim\post_despacho_merged.parquet"
    df_merged.to_parquet(output_path, index=False)
    print(f"\nArchivos combinados correctamente. Se guard√≥ 'post_despacho_merged.parquet'.")
    print(f"El DataFrame resultante tiene {df_merged.shape[0]} filas y {df_merged.shape[1]} columnas.")
    
    # Mostrar el DataFrame final (ten cuidado si es muy grande)
    df_merged
else:
    print("No se encontraron archivos Parquet en la carpeta especificada.")
```

```python
df_merged['FECHA'] = pd.to_datetime(df_merged['FECHA'], format='%Y-%m-%dT%H:%M:%S')
df_merged
```

```python
# Configurar para que se muestren todas las filas
pd.set_option('display.max_rows', None)

# Contar las ocurrencias de cada valor en la columna 'CENTRAL'
central_counts = df_merged['CENTRAL'].value_counts().reset_index()
central_counts.columns = ['CENTRAL', 'count']

# Mostrar el DataFrame resultante
display(central_counts)

# (Opcional) Restaurar la configuraci√≥n predeterminada si es necesario
pd.reset_option('display.max_rows')
```

```python
# Configurar para que se muestren todas las filas
pd.set_option('display.max_rows', None)

# Ordenar el DataFrame por la columna 'CENTRAL' (orden alfab√©tico)
central_counts_sorted = central_counts.sort_values(by='CENTRAL')
# Mostrar el DataFrame ordenado
display(central_counts_sorted)
```

#### Transformaciones

```python
# 1. Filtrar las centrales que quieres borrar
centros_a_borrar = [
    "ARROYO BARRIL", "CAYMAN", "DAJABON", "LOS MINA 1", "LOS MINA 2", 
    "HAINA 3", "PUERTO PLATA 1", "PUERTO PLATA 2", "SANTO DOMINGO 5", 
    "SANTO DOMINGO 8", "TIMBEQUE 1", "TIMBEQUE 2"
]

df_filtrado = df_merged[~df_merged['CENTRAL'].isin(centros_a_borrar)].copy()

# 2. Realizar los reemplazos de nombres en la columna 'CENTRAL'
# Cambiar cualquier variante que empiece con "AES ANDR√âS" o "AES ANDRES" a "AES ANDRES"
df_filtrado['CENTRAL'] = df_filtrado['CENTRAL'].str.replace(
    r'^AES ANDR[√âE]S.*', 
    'AES ANDRES', 
    regex=True
)

# Cambiar cualquier variante que empiece con "GENERACI√ìN DE EMERGENCIA AES ANDR√âS" a "GENERACION DE EMERGENCIA AES ANDRES"
df_filtrado['CENTRAL'] = df_filtrado['CENTRAL'].str.replace(
    r'^GENERACI[√ìO]N DE EMERGENCIA AES ANDR[√âE]S.*', 
    'GENERACION DE EMERGENCIA AES ANDRES', 
    regex=True
)
```

```python
# Filtrar las filas de AES ANDRES
df_aes = df_filtrado[df_filtrado["CENTRAL"] == "AES ANDRES"].copy()

# Definir la lista de columnas de horas (asumiendo que se llaman H1 a H24)
hour_cols = [f"H{i}" for i in range(1, 25)]

# Sumar las columnas de horas en una nueva columna 'total_hours'
df_aes["total_hours"] = df_aes[hour_cols].sum(axis=1)

# Asegurarse de que la columna FECHA est√© en formato datetime
df_aes["FECHA"] = pd.to_datetime(df_aes["FECHA"], errors="coerce")

# Para cada fecha, conservar solo la fila con mayor 'total_hours'
df_aes_unique = df_aes.loc[df_aes.groupby("FECHA")["total_hours"].idxmax()].copy()

# (Opcional) Eliminar la columna auxiliar si ya no la necesitas
df_aes_unique.drop(columns=["total_hours"], inplace=True)

df_aes_unique
```

```python
# Configurar para que se muestren todas las filas
pd.set_option('display.max_rows', None)

# Contar las ocurrencias de cada valor en la columna 'CENTRAL'
central_counts = df_filtrado['CENTRAL'].value_counts().reset_index()
central_counts.columns = ['CENTRAL', 'count']

# Mostrar el DataFrame resultante
display(central_counts)

# (Opcional) Restaurar la configuraci√≥n predeterminada si es necesario
pd.reset_option('display.max_rows')
```
